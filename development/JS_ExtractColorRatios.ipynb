# Extracting data locations and data verification

```python
file_pic = './Data/palettes-dataframe.csv' 
```

```python
import webcolors
import colorsys

df_pic = pd.read_csv(file_pic)

custom_colors = {
    'lilac': (200, 162, 200),
    'rose': (255, 0, 127),
    'teal': (0, 128, 128)
}

def colorname_cleanup(colorname):
    # Convert color name to RGB values
    # Strip whitespace characters from the color name
    colorname = colorname.strip()

    if colorname == 'gery':
      colorname = 'grey'
    elif colorname == 'gret':
      colorname = 'grey'
    elif colorname == 'piml':
      colorname = 'pink'
    elif colorname == 'pirple':
      colorname = 'purple'
    elif colorname == 'purpole':
      colorname = 'purple'
    elif colorname == 'back':
      colorname = 'black'
    elif colorname == 'brwon':
      colorname = 'brown'
    
    return colorname

# Apply the get_hsv function to the DataFrame and create new columns for the HSV values
for i in range(len(df_pic)):
  df_pic['Colour 1'].iloc[i] = colorname_cleanup(df_pic['Colour 1'].iloc[i]) 
  df_pic['Colour 2'].iloc[i] = colorname_cleanup(df_pic['Colour 2'].iloc[i]) 
  df_pic['Colour 3'].iloc[i] = colorname_cleanup(df_pic['Colour 3'].iloc[i]) 
```

```python
df_pic
```

```python
unique_colours = pd.unique(df_pic['Colour 3'])
unique_episodes = pd.unique(df_pic['Episode'])
print(unique_colours)
print(unique_episodes)
```

```python
count = []
for i, episode in enumerate(unique_episodes):
  df_sub = df_pic[df_pic['Episode'] == episode]


  count_C1 = df_sub['Colour 1'].value_counts()
  count_C2 = df_sub['Colour 2'].value_counts()
  count_C3 = df_sub['Colour 3'].value_counts()

  count.append(count_C1 + count_C2 + count_C3)
print(count)
```

```python
df_count = pd.DataFrame(count, columns=[unique_colours])
```

```python
count[1]/sum(count[1])*100

```

```python
df = pd.DataFrame()
for i, episode in enumerate(unique_episodes):
  df1 = count[i].reset_index()
  df_transposed = df1.T.set_axis(df1.index, axis=1, inplace=False)
  df_transposed.columns = df_transposed.iloc[0]
  df_transposed = df_transposed.drop(df_transposed.index[0])
  df_transposed.fillna(0,inplace=True)
  df = pd.concat([df, df_transposed], ignore_index = True)
```

```python
df.fillna(0,inplace=True)
df
```

```python
#Quick check
row_sum = df.sum(axis=1)
df = df.div(row_sum, axis=0)  # divide each element in a row by the sum of that row
df.head()
```
