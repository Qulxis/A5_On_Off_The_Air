# Filter csv files
This notebook cleaned the labeled data and combines it with the exisitng csv files to create different csv combinations for different cases

```python
import pandas as pd
import numpy as np
```

```python
pallets = pd.read_csv("./palettes-dataframe.csv")
urls = pd.read_csv('testing2b.csv')
pallets.head()
```

```python
#clean pallets
pallets = pallets.rename(columns={"Colour 1": "Colour1","Colour 2":"Colour2", "Colour 3":"Colour3"})
print(pallets.head())
all_comb = set()
all_comb.update(list(pallets.Colour1.unique()))
all_comb.update(list(pallets.Colour2.unique()))
all_comb.update(list(pallets.Colour3.unique()))
# print(pallets.Colour2.unique())
all_comb #See all the possible mis lableing
```

```python
#Create mapping to update:
update_mapping = {
    'black ': 'black',
    'blue ':'blue',
    'gery':'grey',
    'green ':'green',
    'gret':'grey'
}
pallets['Colour1'] = pallets['Colour1'].replace(update_mapping)
pallets['Colour2'] = pallets['Colour2'].replace(update_mapping)
pallets['Colour3'] = pallets['Colour3'].replace(update_mapping)

```

```python
#check all unique labels agains:

all_comb = set()
all_comb.update(list(pallets.Colour1.unique()))
all_comb.update(list(pallets.Colour2.unique()))
all_comb.update(list(pallets.Colour3.unique()))
all_comb #See all the possible mis lableing
```

Looks good, now we can save the csv file however we wish

```python
# Looks like we need to rename urls colums nnames
urls = urls.rename(columns={"Category":"Episode"})
urls.head()
```

```python
#Example merdge:
result_df = pd.merge(pallets, urls, on=['Episode', 'Frame'], how='inner')
result_df.head()
```
